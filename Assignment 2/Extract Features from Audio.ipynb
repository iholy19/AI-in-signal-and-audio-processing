# Import Libraries

```python
import csv, os
import librosa as lb
import numpy as np
```

```python
header = 'filename chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header += ' label'
header = header.split()
```

# Extract audio features from the input dataset

Ref: In my bachelors, I worked with audio processing. The same code is used here to extract features from the audio

```python
audio_dataset_path = r'drive/MyDrive/EPITA/Signal Processing/Assi 2/dataset_words_test.csv'
file = open(audio_dataset_path, 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)

categories = 'yes no wow left right'.split()
for g in categories:
    for filename in os.listdir(f"./drive/MyDrive/EPITA/Signal Processing/Assi 2/data/audio_test/{g}"):
        audio = f"./drive/MyDrive/EPITA/Signal Processing/Assi 2/data/audio_test/{g}/{filename}"
        y, sr = lb.load(audio, mono=True, duration=15)
        rmse = lb.feature.rms(y=y)
        chroma_stft = lb.feature.chroma_stft(y=y, sr=sr)
        spec_cent = lb.feature.spectral_centroid(y=y, sr=sr)
        spec_bw = lb.feature.spectral_bandwidth(y=y, sr=sr)
        rolloff = lb.feature.spectral_rolloff(y=y, sr=sr)
        zcr = lb.feature.zero_crossing_rate(y)
        mfcc = lb.feature.mfcc(y=y, sr=sr)
        to_append = f'{filename} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'
        for e in mfcc:
            to_append += f' {np.mean(e)}'
        to_append += f' {g}'
        file = open(audio_dataset_path, 'a', newline='')
        with file:
            writer = csv.writer(file)
            writer.writerow(to_append.split())
```
